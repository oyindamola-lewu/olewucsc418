```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import cv2 as cv
```

```python
img = cv.imread('img/sst.jpg', 0)
rows, cols = img.shape
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
```

```python
M = np.float32([[1, 0, 100],[0, 1, 50]])
trans1 = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(trans1)
```
